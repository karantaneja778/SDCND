<b> Hough Transform to Find Lane Lines </b>

```python
# imports
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import cv2
```

```python
# Read in and grayscale the image
image = mpimg.imread('exit-ramp.jpg')

gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap = 'Greys_r')
```

```python
# Define a kernel size and apply gaussian smoothing
kernel_size = 5

blur_gray = cv2.GaussianBlur(gray, (kernel_size, kernel_size), 0)

plt.imshow(blur_gray, cmap = 'Greys_r')
```

```python
# Define parameters for Canny and apply
low_threshold = 100
upper_threshold = 200

edges = cv2.Canny(blur_gray, low_threshold, upper_threshold)

plt.imshow(masked_edges, cmap = 'Greys_r')
```

```python
mask = np.zeros_like(edges)   
ignore_mask_color = 255  

# Define four sided polygon to mask
imshape = image.shape
vertices = np.array([[(0,imshape[0]),(imshape[1]/2, imshape[0]/2), (imshape[1]/2, imshape[0]/2), (imshape[1],imshape[0])]], dtype=np.int32)
cv2.fillPoly(mask, vertices, ignore_mask_color)
masked_edges = cv2.bitwise_and(edges, mask)
```

```python
# Define the Hough transform parameters
rho = 2
theta = np.pi/180
threshold = 15
min_line_length = 40
max_line_gap = 20

# make a blank copy of the same size image to draw on
# creating a blank to draw lines
line_image = np.copy(image)*0
```

```python
# Run Hough Transform on edge detected image
lines = cv2.HoughLinesP(masked_edges, rho, theta, threshold, np.array([]), min_line_length, max_line_gap)

# Iterate over the outpt lines and draw lines on the blanks
for line in lines:
    for x1, y1, x2, y2 in line:
        cv2.line(line_image, (x1, y1), (x2, y2), (255, 0, 0), 10)
        
# Create a color binary image to combine with the line image
color_edges = np.dstack((masked_edges, masked_edges, masked_edges))

# Draw the lines on the edge image
combo = cv2.addWeighted(color_edges, 0.8, line_image, 1, 0)
plt.imshow(combo)
```
